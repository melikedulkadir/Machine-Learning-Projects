With the help of the load_images function, the training and test data were loaded and the images were resized according to the given parameters.

```python
def load_images(path,size):
    folders= os.listdir(path)
    data = []
    label = []
    for i in folders:
        images=os.listdir(path+'/'+i)
        if '.DS_Store' in images:
            images.remove('.DS_Store')
        for j in images:
            img =Image.open(path+'/'+i+'/'+j)
            img = ImageOps.grayscale(img)
            img=img.resize((size,size), image.NEAREST)
            ar=np.array(img)
            data.append(ar)
            label.append(i)
    t=pd.factorize(np.array(label))
    tgt=t[0]
    return np.array(data), tgt
```

### 1. Extract features for each image in training set

Shape is an important and powerful feature for regression process on images. Edge information in the image is obtained by using the Canny edge detection. This function is used to apply Canny to every picture in the training data.

```python
def apply_canny(data,t_lower,t_upper):
    canny_images = []
    for i in range(len(data)):
        canny_img = cv2.Canny(data[i], t_lower, t_upper)
        canny_images.append(canny_img)
    return canny_images
```

```python
canny_train_images = apply_canny(train_x,100,250)
canny_test_images = apply_canny(test_x,100,250)
```

The Gabor filter is a commonly used filter for texture extraction. At this stage, the necessary preprocesses were applied first, then the Gabor filter was applied to each image in the training.

```python
from multiprocessing.pool import ThreadPool

def build_filters():
    filters = []
    ksize = 31
    for theta in np.arange(0, np.pi, np.pi / 16):
        kern = cv2.getGaborKernel((ksize, ksize), 4.0, theta, 10.0, 0.5, 0.8, ktype=cv2.CV_32F)
        kern /= 1.5*kern.sum()
        filters.append(kern)
    return filters

def process(img, filters):
    accum = np.zeros_like(img)
    for kern in filters:
        fimg = cv2.filter2D(img, cv2.CV_8UC3, kern)
        np.maximum(accum, fimg, accum)
    return accum

def process_threaded(img, filters, threadn = 8):
    accum = np.zeros_like(img)
    def f(kern):
        return cv2.filter2D(img, cv2.CV_8UC3, kern)
    pool = ThreadPool(processes=threadn)
    for fimg in pool.imap_unordered(f, filters):
        np.maximum(accum, fimg, accum)
    return accum
```

```python
def apply_gabor(data):
    filters = build_filters()
    filtered = []
    for i in range(len(data)):
        res1 = process(data[i], filters)
        res2 = process_threaded(data[i], filters)
        if (res1 == res2).all():
            filtered.append(res2)
    return filtered
```

```python
filtered_train_images = apply_gabor(canny_train_images)
filtered_test_images = apply_gabor(canny_test_images)
```

### 2. KNN and WeightedKNN

The euclidean_distance function is the function that measures the distance between two samples required to apply KNN.

```python
def euclidean_distance(first_sample, second_sample):
    if len(first_sample) != len(second_sample):
        raise Exception("Two rows must have the same dimension!")
    distance = np.sum((first_sample-second_sample)**2)
    return distance
```

It is the KNN algorithm created to make classification with knn. It takes k as a parameter. It ensures that the correct class is predicted with the fit, predict, vote functions.

In the vote function here, the calculation is made by accepting the votes of all neighbors as equal.

```python
from collections import Counter
class Knn:

    # setting k value of KNN algorithm
    def __init__(self, k=3):
        self.k = k

    # setting attribute and class data
    def fit(self, X, y):
        self.X_train = X
        self.y_train = y

    def predict(self, X_test):
        predicted_classes = []
        for test in X_test:
            distances = [[euclidean_distance((test).flatten(), (self.X_train[i]).flatten()), i] for i in range(len(self.X_train))]
            predicted= self.vote(sorted(distances, key=lambda x: x[0])[:self.k])
            predicted_classes.append(predicted)
        return predicted_classes

    def vote(self, neighbours):
        distances = []
        classes = []
        indexes = []

        for neighbour in neighbours:
            index = neighbour[1]  # index of related data's row

            true_class = self.y_train[index]  # true label for the corresponding data
            distance = neighbour[0]  # distance

            # adding information into lists
            distances.append(distance)
            classes.append(true_class)
            indexes.append(index)

        freq_array = Counter(classes).most_common()
        most_common = freq_array[0] # find the most repetitive class

        predicted_class = int(most_common[0])

        return predicted_class
```

```python
class WeightedKNN():

    # setting k value of KNN algorithm
    def __init__(self, k=3):
        self.k = k

    def fit(self, X, y):
        self.X_train = X
        self.y_train = y

    def predict(self, X_test):
        predicted_classes = []
        for test in X_test:
            distances = [[euclidean_distance(test.flatten(), (self.X_train[i]).flatten()), i] for i in range(len(self.X_train))]
            sorted_array = sorted(distances, key=lambda x: x[0])
            weight_dict = {(i): 0 for i in range(3)} # distance storing
            label_dict = {(j): 0 for j in range(3)} # repetition storing
            for dist in range(self.k):
                neighbour_label = self.y_train[sorted_array[dist][1]]
                if not sorted_array[dist][0] == 0:
                    weight_dict[neighbour_label] += (1 / sorted_array[dist][0]) # add voting value
                    label_dict[neighbour_label] += 1 # increase the number of being seen for each class type
                else:
                    # if distance value equals to 0, weight value becomes 1 to avoid zero division error
                    weight_dict[neighbour_label] += 1
                    label_dict[neighbour_label] += 1

            closest_ones_array = [(label, weight_dict[label]) for label in weight_dict.keys()]
            closest_ones_array = sorted(closest_ones_array, key=lambda x: x[1], reverse=True)

            label_freq_array =  [(label, label_dict[label]) for label in label_dict.keys()]
            label_freq_array = sorted(label_freq_array, key=lambda x: x[1], reverse=True)

            closest_one = closest_ones_array[0]
            most_repeating = label_freq_array[0]

            repeating_classes = []
            vals=list(weight_dict.values())
            for key, value in weight_dict.items():
                if vals.count(value) > 1:
                    repeating_classes.append(key)

            if len(repeating_classes) > 0:
                predicted = int(most_repeating[0])
                predicted_classes.append(predicted)
            else:
                predicted = int(closest_one[0])
                predicted_classes.append(predicted)

        return predicted_classes
```

Weighted kNN is a modified version of the k nearest neighbor. Another issue is the approach to combining class labels. The simplest method is to get a majority vote, but this can be a problem if the nearest neighbors' distances vary greatly and the nearest neighbors more reliably represent the object's class.

It is the Weighted KNN algorithm created to make classification with knn. It takes k as a parameter. It ensures that the correct class is predicted with the fit, predict, vote functions. In vote function not all the neighbors votes equeal.


Applying the knn and weighted knn algorithm, makes predictions and printing the results.

```python
from sklearn.metrics import confusion_matrix

def apply_knn(X_train,X_test,test_y,title):
    for i in range(3,8,2):
        model = Knn(k=i)
        model.fit(X_train, train_y)
        predictions =  model.predict(X_test)
        cm = confusion_matrix(test_y, predictions)
        cm_df = pd.DataFrame(cm, index = ["covid","normal","pneumonia"], columns = ["covid","normal","pneumonia"])
        print(cm_df)
        print()
        print("Classification Report for",title,"for k= {}".format(i))
        print()
        print(classification_report(test_y, predictions,target_names=["covid","normal","pneumonia"]))
```

```python
def apply_weighted_knn(X_train,X_test,test_y,title):
    for i in range(3,8,2):
        model = WeightedKNN(k=i)
        model.fit(X_train, train_y)
        predictions =  model.predict(X_test)
        cm = confusion_matrix(test_y, predictions)
        cm_df = pd.DataFrame(cm, index = ["covid","normal","pneumonia"], columns = ["covid","normal","pneumonia"])
        print(cm_df)
        print()
        print("Classification Report for",title,"for k= {}".format(i))
        print()
        print(classification_report(test_y, predictions,target_names=["covid","normal","pneumonia"]))
```

```python
from PIL import Image
import torch
import torchvision.models as models
import torchvision.transforms as transforms
from torch.autograd import Variable
import os
print(os.listdir("train"))
```

After importing the necessary libraries, the model was created using models.resnet18.

```python
model = models.resnet18(pretrained='imagenet')
```

ResNet-18 expects images to be at least 224x224 and also normalized with a given mean and standard deviation. So first defined some PyTorch transformations:

```python
#Resize the image to 224x224 px
scaler = transforms.Resize((224, 224))
normalize = transforms.Normalize(mean=[0.485, 0.456, 0.406],
                                 std=[0.229, 0.224, 0.225])
to_tensor = transforms.ToTensor()

layer = model._modules.get('avgpool')
```

The layer was created to create a reference to the layer we want to remove. The first layers in the network usually learn high-level features such as 'image contains fur' or 'image contains round object', while lower-level features are more specific to the training data. The 'avgpool' layer selected here is at the end of ResNet-18.

In this step, the 'get_vector' function was created using all the steps. This function takes an image path and returns a PyTorch tensor representing the properties of the image.

```python
def extract_feature_vector(img):

    t_img = Variable(normalize(to_tensor(scaler(img))).unsqueeze(0))
    my_embedding = torch.zeros(1, 512, 1, 1)

    def copy_data(m, i, o):
        my_embedding.copy_(o.data)

    h = layer.register_forward_hook(copy_data)

    model(t_img)

    h.remove()

    return my_embedding.squeeze().numpy()
```

Opening the images sequentially and calculating the cosine distance.

```python
def get_image(image):
    return Image.open(image).convert('RGB')
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def get_cosine_distance(image1,image2):
    im1, im2 = get_image(image1), get_image(image2)
    image1 = extract_feature_vector(im1).reshape(1, -1)
    image2 = extract_feature_vector(im2).reshape(1, -1)
    return cosine_similarity(image1, image2)
```
